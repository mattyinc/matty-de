```python
import duckdb
import pandas as pd

# Create a simple dataframe
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie'],
    'score': [95, 82, 78]
})

# Query it with SQL
result = duckdb.query("SELECT name, score FROM df WHERE score > 80 ORDER BY score DESC").df()
print(result)
```

```python
result = duckdb.query("SELECT * FROM df").df()
print(result)
```
