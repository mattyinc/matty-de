## Data Quality Findings

After exploring the dataset, I identified 3 issues to clean:

1. **passenger_count = 0** — trips with no passengers are invalid records
2. **trip_distance = 0** — a taxi going nowhere is not a real trip  
3. **fare_amount = 0** — zero fare trips are errors or test entries

These will be filtered out in the next notebook: `02_data_cleaning.ipynb`
